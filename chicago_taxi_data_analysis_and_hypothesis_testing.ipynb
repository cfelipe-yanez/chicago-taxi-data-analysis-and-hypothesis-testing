# Paso 5. Prueba de hipótesis (Python)

```python
# Ruta del archivo
file_path = "/datasets/project_sql_result_07.csv"

# Leer el archivo CSV
df_final = pd.read_csv(file_path)

# Mostrar las primeras filas
print(df_final.head(10))
```

### HIPOTESIS
La duración promedio de los viajes es igual entre sábados con buen clima y sábados con mal clima

### NULA
La duración promedio de los viajes es diferente en sábados con buen clima vs sábados con mal clima.

### Alfa = 5%

Si p-valor < 0.05, rechazamos H₀ → el clima (lluvioso) afecta la duración.

Si p-valor ≥ 0.05, no hay evidencia suficiente para rechazar H₀.

```python
from scipy import stats

# Filtrar solo sábados
df_final["start_ts"] = pd.to_datetime(df_final["start_ts"])
df_final["day_of_week"] = df_final["start_ts"].dt.day_name()

saturdays = df_final[df_final["day_of_week"] == "Saturday"]

# Separar en grupos por condición climática
good_weather = saturdays[saturdays["weather_conditions"] == "Good"]["duration_seconds"]
bad_weather  = saturdays[saturdays["weather_conditions"] == "Bad"]["duration_seconds"]

# Prueba de hipótesis (t-test para muestras independientes)
t_stat, p_value = stats.ttest_ind(good_weather, bad_weather, equal_var=False)

print("t-statistic:", t_stat)
print("p-value:", p_value)

alpha = 0.05
if p_value < alpha:
    print("Rechazamos H0: el clima afecta la duración promedio.")
else:
    print("No hay evidencia suficiente para decir que el clima afecta la duración.")
```

<div class="alert alert-block alert-success">
<b>Comentario de Revisor          </b> <a class="tocSkip"></a>

Bien hecho con el test de hipótesis, fue realizado de forma correcta. 

Es importante mencionar que el t-test asume que las dos muestras tienen varianzas iguales. Para verificar esa igualdad de varianzas podemos usar el **test de Levene**. Este nos dice si la variabilidad entre los grupos es estadísticamente diferente, podrías usarlo en un siguiente test. En corto, el flujo sería así

1. Usamos `levene()` para verificar si las varianzas son iguales.
2. Dependiendo del resultado, usamos `ttest_ind()` con el argumento `equal_var`:
   - `equal_var=True` si Levene dice que las varianzas son iguales (p > 0.05).
   - `equal_var=False` si Levene dice que las varianzas son diferentes (p < 0.05).





</div>

### Planteamiento de las hipótesis

- Hipótesis nula (H₀): la duración promedio de los viajes del Loop al aeropuerto O’Hare es la misma los sábados con buen clima y los sábados con mal clima.

- Hipótesis alternativa (H₁): la duración promedio de los viajes es diferente en sábados con buen clima frente a sábados con mal clima.

### Criterio de prueba

Se fijó un nivel de significancia de α = 0.05, valor comúnmente utilizado en pruebas estadísticas.

Para comparar las medias de los dos grupos (sábados con clima “Good” y sábados con clima “Bad”) se aplicó una prueba t de Student para dos muestras independientes, ya que se trata de una variable continua (tiempo en segundos) y dos poblaciones distintas.

### Resultados e interpretación

- La prueba arrojó un t-statistic = -7.19 y un p-value ≈ 6.7e-12, el cual es mucho menor que 0.05.

Por lo tanto, se rechaza la hipótesis nula y se concluye que las condiciones climáticas sí afectan significativamente la duración promedio de los viajes del Loop al aeropuerto O’Hare en los sábados de noviembre de 2017.
